```python
from time import sleep
import selenium
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import re
```

```python
# type the job title and area in the search box and copy the url here.
url = "https://www.linkedin.com/jobs/search/?currentJobId=656674657&f_LF=f_AL&keywords=data%20scientist&location=San%20Francisco%20Bay%20Area&locationId=us%3A84"
driver = webdriver.Chrome("E:\Application\chromedriver")
driver.get(url)

user = 'yourusername'
pas = 'yourpassword'
```

```python
#navigate to the login in page
login = driver.find_element_by_xpath('//*[@id="mobile-chrome"]/nav/div/ul/li[1]/a')
login.click()

#login
username =driver.find_element_by_xpath('//*[@id="session_key-login"]')
username.clear()
username.send_keys(user)

password = driver.find_element_by_xpath('//*[@id="session_password-login"]')
password.clear()
password.send_keys(pas)

log = driver.find_element_by_xpath('//*[@id="btn-primary"]')
log.click()
```

```python
ele_pager = driver.find_element_by_css_selector('.page-list ol')   
max_page = int(ele_pager.text[-1])
firstpage = driver.current_url

for i in range(1,max_page+1):
    try:
        if i >1:
            page = '&start='
            p=25*(i-1)
            url = firstpage+page+str(p)
            driver.get(url)

        for j in range(1,25):
            ttl_app = i*j
            before = '.ember-view:nth-child('
            after = ') .pv3 div'
            add = ''.join([before, str(j),after])
            #click each card
            try:
                ele = driver.find_element_by_css_selector(css_selector=add) 
                ele.click()
            except WebDriverException:
                continue
            sleep(2)
            #try cath for if applied
            try:
                driver.find_element_by_css_selector('.jobs-s-apply__button')
            except NoSuchElementException:
                print("Already applied")
                continue

            #if try successful,click easy-apply button
            easyapply = driver.find_element_by_css_selector('.jobs-s-apply__button')
            easyapply.click()
                       
            #check whether redirect to another page or apply instantly
            if (len(driver.window_handles) == 2):
                driver.switch_to.window(window_name=driver.window_handles[-1])
                sleep(2)
                try:
                    submit = driver.find_element_by_css_selector('.continue-btn')
                    submit.click()
                    print(str(ttl_app)," resume submitted")
                except:
                    print('cant apply in second page')
                driver.close()
                driver.switch_to.window(window_name=driver.window_handles[0])
            #click choose resume button
            else:
                try:
                    ele_resume = driver.find_element_by_css_selector(".jobs-apply-form__upload-options")
                    ele_resume.click()
                    #choose first resume
                    ele_choose_resume = driver.find_element_by_css_selector(".jobs-apply-form__recent-resume-filename")
                    ele_choose_resume.click()
                    #click submit button
                    ele_submit_resume = driver.find_element_by_css_selector(".jobs-apply-form__submit-button.button-primary-large")
                    ele_submit_resume.click()
                    print(str(ttl_app)," resume submitted")
                    end_page = driver.find_element_by_css_selector(".modal-content-wrapper")
                    actions = webdriver.common.action_chains.ActionChains(driver)
                    action = actions.move_to_element_with_offset(end_page, -20, 0)
                    action.click()
                    action.perform()
                    # check if there is a refererr:
                    try: 
                        refer = driver.find_element_by_css_selector('.js-post-apply-ask-for-referral-button')
                        refer.click()
                        send = driver.find_element_by_css_selector('.js-submit-referral-message')
                        send.click()
                    except:
                        print('there was an error in reference or there is no reference')
                        continue
                except: 
                    print("can't apply in the same page")
            sleep(1)
    except:
        print('sth is wrong')
            
    sleep(5)


 
```
